## Wikipedia
First, we'll practice using `pywikibot` to explore Wikipedia. Wikipedia contains a bunch of pages that link to each other, so it is in essence a directed graph; one [interesting fact](https://en.wikipedia.org/wiki/Wikipedia:Getting_to_Philosophy) about Wikipedia is that if you click the first link on a page that is *not* in italics or parentheses, you will very likely either go in an infinite loop or end up at Philosophy. Today, we're going to practice doing random walks of Wikipedia. 

```python
import pywikibot
site = pywikibot.Site('en', 'wikipedia')
orca_page = pywikibot.Page(site, 'Orca')
linked_pages = [p for p in orca_page.linkedPages()]
```
